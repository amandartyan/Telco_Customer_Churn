```python
plt.figure(figsize=(12,6))

features = ['tenure', 'MonthlyCharges', 'TotalCharges', 'SeniorCitizen']
for i in range (0, len(features)):
  plt.subplot(2, (len(features)//2) + 1, 1 + i)
  sns.kdeplot(x=df[features[i]], color='blue')
  plt.xlabel(features[i])
  plt.tight_layout()

plt.show()
```

## Observation:


*   The continuous numerical features, are a somewhat positively skewed.
*   Senior citizen have the positively skewed too

*   monthly charges seems like have 2 peaks, it can be indicating as bimodal distribution. And the imbalance is still ok
